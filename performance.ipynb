## Display Reward function

```python
def display_rewards(episodes: int, rewards: list, title: str):
    """
    displays rewards throughout training
    """
    window_size = 1
    moving_avg = np.convolve(rewards, np.ones(window_size)/window_size, mode='valid')
    
    # plt.plot(range(1, episodes+1), rewards, color="mediumorchid", label="rewards")
    # plt.plot(range(1, episodes+1), [np.mean(rewards)]*episodes, color="red", label="mean_reward")
    plt.plot(range(window_size, episodes+1), moving_avg, color="blue", label="moving_avg")
    
    plt.xlabel("Episodes")
    plt.ylabel("Rewards")
    plt.title(f"{title} - Mean reward: {np.mean(rewards):.2f}")
    plt.legend()
    plt.show()
```

## Random Agent

```python
def test_random_agent(env, num_games=10):
    
    total_wins = 0
    total_draws = 0
    total_loss = 0
    for game in range(num_games):
        state, player = env.reset()
        done = False
        game_reward = 0.0
        while not done:
            if player == env.WHITE_PAWN:
                valid_moves = env.get_available_moves(state, env.WHITE_PAWN)
                if not valid_moves:
                    break
                move = random.choice(valid_moves)
                state, reward, done, next_player = env.step(move)
                game_reward += reward
            else:
                valid_moves = env.get_available_moves(state, env.BLACK_PAWN)
                if not valid_moves:
                    break
                move = random.choice(valid_moves)
                state, reward, done, next_player = env.step(move)
                game_reward += reward
            # env.render()
            # time.sleep(0.3)
            player = next_player

        # Check the game's outcome (win, loss, or draw)
        _, winner = env.is_game_finished(env.current_state)
        if winner == env.WHITE_PAWN:
            total_wins += 1
            print(f"Game {game+1}: Total Reward = {game_reward}, Winner = Agent")
        elif winner == env.BLACK_PAWN:
            total_loss += 1
            print(f"Game {game+1}: Total Reward = {game_reward}, Winner = Opponent")
        else:
            total_draws += 1
            print(f"Game {game+1}: Total Reward = {game_reward}, Draw")
    
    return total_wins,total_draws,total_loss

wins,draws,losses = test_random_agent(env, num_games=100)
wr_rd = wins/(wins+losses+draws)
lr_rd = losses/(wins+losses+draws)
dr_rd = draws/(wins+losses+draws)

print("wins : ",wr_rd," | draws : ",dr_rd," | losses : ",lr_rd)
```

## Training Loop

```python
def train_agent(num_episodes=250):
    agent = MonteCarloAgent(epsilon=1.0, epsilon_min=0.01, epsilon_decay=0.999, gamma=0.906)
    scores = []
    max_reward = -np.inf
    q_variances = []
    
    progress_bar = tqdm(range(num_episodes), desc="Episodes", unit="iteration")

    for episode in progress_bar:
        if episode != 0:
            
            agent = torch.load("best_model_MC.pth")
            agent.epsilon = new_epsilon
        state, player = env.reset()
        episode_data = []
        done = False
        total_reward = 0.0
        prev_state = deepcopy(state)
        while not done:
            if player == env.WHITE_PAWN:
                # Agent's turn: choose a move and update reward
                move, action_index = agent.policy(state)
                if move is None:
                    break
                new_state, reward_env, done, next_player = env.step(move)
                reward_enriched = compute_enriched_reward(prev_state, new_state, move, env, player=env.WHITE_PAWN)
                reward = reward_env + reward_enriched
                episode_data.append((state, action_index, reward))
                total_reward += reward
            else:
                # Opponent's turn: select a random move
                valid_moves = env.get_available_moves(state, env.BLACK_PAWN)
                if not valid_moves:
                    break
                move = random.choice(valid_moves)
                new_state, reward_env, done, next_player = env.step(move)
                total_reward += reward_env
            prev_state = deepcopy(state)
            state = new_state
            player = next_player

        # Store the episode score and update the agent's knowledge
        scores.append(total_reward)
        agent.update(episode_data)
        new_epsilon = agent.update_epsilon()

        # Calculate and track Q-value variance to monitor agent's learning progress
        q_vals = []
        for state_key in agent.Q:
            q_vals.extend(agent.Q[state_key])
        q_variance = np.var(q_vals) if q_vals else 0.0
        q_variances.append(q_variance)
        
        
        if total_reward >= max_reward:
            torch.save(agent, "best_model_MC.pth")
            
            print("New best",total_reward)
            max_reward = total_reward
        else:
            progress_bar.set_postfix(total_reward=total_reward,epsilon=agent.epsilon,New_best="not new best")
            
    return agent, scores,q_variances
```

## Test Loop

```python
def test_agent(agent, env, num_games=10):
    old_epsilon = agent.epsilon
    agent.epsilon = 0.0  # Set epsilon to 0 for testing (no exploration)
    total_wins = 0
    total_draws = 0
    total_loss = 0
    games_history = []
    for game in range(num_games):
        state, player = env.reset()
        done = False
        game_reward = 0.0
        game_history = []
        while not done:
            game_history.append(state)
            if player == env.WHITE_PAWN:
                # Agent's turn: choose a move using the learned policy
                move, action_index = agent.policy(state)
                if move is None:
                    break
                state, reward, done, next_player = env.step(move)
                game_reward += reward
            else:
                # Opponent's turn: select a random move
                valid_moves = env.get_available_moves(state, env.BLACK_PAWN)
                if not valid_moves:
                    break
                move = random.choice(valid_moves)
                state, reward, done, next_player = env.step(move)
                game_reward += reward
            # env.render()
            # time.sleep(0.3)
            player = next_player
        _, winner = env.is_game_finished(env.current_state)
        if winner == env.WHITE_PAWN:
            total_wins += 1
            print(f"Game {game+1}: Total Reward = {game_reward}, Winner = MC Agent")
        elif winner == env.BLACK_PAWN:
            total_loss += 1
            print(f"Game {game+1}: Total Reward = {game_reward}, Winner = opponent")
        else:
            total_draws += 1
            print(f"Game {game+1}: Total Reward = {game_reward}, Draw")
        games_history.append(game_history)
    agent.epsilon = old_epsilon
    return total_wins,total_draws,total_loss,games_history
```

```python
def compute_state_feature(state, env):
    """
    Extracts a (row, col) pair representing the average position of the white pieces in the state.
    If no white pieces are present, the center of the board is returned.
    """
    positions = []
    for i in range(env.BOARD_SIZE):
        for j in range(env.BOARD_SIZE):
            if state[i][j] in (env.WHITE_PAWN, env.WHITE_KING):
                positions.append((i, j))
    if len(positions) == 0:
        return (env.BOARD_SIZE // 2, env.BOARD_SIZE // 2)
    avg_row = int(np.mean([pos[0] for pos in positions]))
    avg_col = int(np.mean([pos[1] for pos in positions]))
    return (avg_row, avg_col)

def plot_state_value_function(agent, env):
    """
    Displays a heatmap of the state values, using
    the center of mass of the white pieces as the state feature.
    For each state in the agent's Q-table, we calculate V(s)=max(Q(s,·)).
    Then, these values are grouped by the average position (row, col) of the white pieces.
    """
    grid = np.zeros((env.BOARD_SIZE, env.BOARD_SIZE))
    counts = np.zeros((env.BOARD_SIZE, env.BOARD_SIZE))
    
    for state, q_list in agent.Q.items():
        V = max(q_list)
        f_row, f_col = compute_state_feature(state, env)
        grid[f_row, f_col] += V
        counts[f_row, f_col] += 1
    
    avg_grid = np.divide(grid, counts, out=np.zeros_like(grid), where=counts != 0)
    
    plt.figure(figsize=(6,6))
    plt.imshow(avg_grid, origin='lower', cmap='inferno', interpolation='nearest')
    plt.colorbar(label='Average State Value V(s)')
    plt.title('State-Value Function Heatmap\n(Feature: Average White Piece Position)')
    plt.xlabel('Column')
    plt.ylabel('Row')
    plt.xticks(np.arange(env.BOARD_SIZE))
    plt.yticks(np.arange(env.BOARD_SIZE))
    plt.show()

plot_state_value_function(best_agent, env)
```
